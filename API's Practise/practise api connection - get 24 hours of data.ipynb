```python
def get_24_hour_data(view_id):
    conn = GoogleAnalyticsQuery(secrets='client_secret.json', token_file_name='analytics.dat')
    
    metrics = ['ga:sessions']
    dimensions = ['ga:date','ga:hour','ga:minute']
    
    df = pd.DataFrame()
    
    query = {
    'ids' : 'ga:' + view_id,
    'metrics' : metrics,
    'dimensions' : dimensions,    
    'start_date' : 'yesterday',
    'end_date': 'today',
    'max_results': 100000
    }

    df_holder, metadata = conn.execute_query(**query)
    
    
#     json_data = df_holder.to_json()
    
    return df_holder
```

```python
results = get_24_hour_data('91998326')
#x yyyy-mm-dd hh:mm y sessions
results['x'] = results['date'] + results['hour'] + results['minute']
del results['date']
del results['hour']
del results['minute']
results['x'] = pd.to_datetime(results['x'])
results.columns = ['x', 'y']
results['y'] = results['y'].apply(lambda x: x.strftime('%Y-%m-%d %H:%M'))

```
